```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_4taDosis.csv")
pdosis=pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_4taDosis.csv")
```

```python
pdosis.head()
```

```python
pdosis
```

```python
RM=pdosis[pdosis['Region']=='Metropolitana']
RM
```

```python
col_list=list(RM)
col_list.remove('Codigo region')
col_list.remove('Codigo comuna')
col_list.remove('Poblacion')
col_list.remove('Region')
col_list.remove('Comuna')



RM=pdosis[pdosis["Region"].str.contains("Metropolitana")]
RM=RM.drop([140],axis=0)



RM["Total"] = RM[col_list].sum(axis=1)
```

```python
RM=RM[['Comuna','Poblacion','Total']]
RM
```

```python
fig=plt.subplots(figsize=(20,14))
p1= plt.bar(RM["Comuna"],RM["Total"])
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(RM["Comuna"],RM["Total"])
```

```python
max(RM["Total"])
```

```python
RM["Proporcion"]=RM.Total/RM.Poblacion
```

```python
fig=plt.subplots(figsize=(20,14))
p1= plt.bar(RM["Comuna"],RM["Proporcion"])
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xticks(rotation=90)
plt.show()
```
